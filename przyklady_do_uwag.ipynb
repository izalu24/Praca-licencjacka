```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rho = np.array([10, 10, 20, 20])
x = np.array([0, 1, 1, 2])
plt.title("Gęstość", fontweight="bold")
plt.xlabel("x", fontweight="bold")
plt.plot(x, rho, color="#FF3083", linewidth=3)
plt.ylim(ymin=0)
plt.show()
```

```python
x = np.arange(0, 2, 0.01)
cent = np.arange(0.125, 1.876, 0.25)
rho = 2*np.arctan(10*x - 10) + 10
lines = 2*np.arctan(10*cent - 10) + 10
plt.title("Gęstość", fontweight="bold")
plt.xlabel("x", fontweight="bold")
plt.plot(x, rho, color="#FF3083", linewidth=3)
plt.vlines(x = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2],
            ymin = 0, ymax = [*lines,lines[-1]],
		colors = 'purple', linestyles="--")
plt.hlines(y=lines, xmin=np.arange(0, 2, 0.25), xmax=np.arange(0.25, 2.1, 0.25), color='purple')
plt.ylim(ymin=0)
plt.show()
```
